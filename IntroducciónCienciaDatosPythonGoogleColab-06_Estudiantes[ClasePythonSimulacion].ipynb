*Un numero aleatorio es una variable aleatoria que sigue la distribuci√≥n uniforme en el intervalo $(0,1)$.*

##**Generadores de N√∫meros Aleatorios**

Para los experimentos de simulaci√≥n generalmente se requiere de secuencias grandes de n√∫meros aleatorios, no solamente unos pocos. Los n√∫meros aleatorios generados en esta secuencia deben ser independientes.

2.   Los n√∫meros generados deben ser estad√≠sticamente independientes.

El per√≠odo de un generador de n√∫meros aleatorios indica la cantidad de n√∫meros aleatorios que se pueden generar en una secuencia antes de que estos comiencen a repetirse. El per√≠odo est√° asociado con la reproducibilidad. Si las secuencias no son reproducibles, entonces no puede hablarse de per√≠odo. 5.   Eficiencia. El m√©todo debe requerir el m√≠nimo tiempo para generar un n√∫mero aleatorio, es decir, debe generar n√∫meros a grandes velocidades.
6.   El m√©todo debe requerir una m√≠nima capacidad de almacenamiento.
7.   Portabilidad. El m√©todo debe ser tal que se pueda transportar de una m√°quina a otra, sin ninguna complicaci√≥n.

Las dos primeras condiciones est√°n asociadas con las condiciones b√°sicas de un generador de n√∫meros aleatorios: Uniformidad e independencia. Si un m√©todo cumple estas dos condiciones, se lo puede considerar como un generador de n√∫meros aleatorios.

En algunos estudios de simulaci√≥n se requieren miles y a veces millones de n√∫meros aleatorios, y debemos garantizar que la mayor parte del tiempo de la simulaci√≥n no se utilice en la generaci√≥n de los n√∫meros sino en los otros procesos del sistema.

El primer paso consiste en definir el m√©todo a usar para generar los n√∫meros
El segundo consiste en definir el conjunto o bater√≠a de pruebas estad√≠sticas a que se debe someter el generador para garantizar las propiedades estad√≠sticas de los n√∫meros generados.

Para generar n√∫meros aleatorios se pueden usar diferentes procedimientos, siempre y cuando garanticen que se cumplen las condiciones b√°sicas de uniformidad e independencia. La exigencia de las dem√°s condiciones depende del uso que se le vaya a dar al generador. ##**Simulaci√≥n est√°tica o sin eje de tiempo**

Se usa para modelar fen√≥menos probabil√≠sticos que no dependen del tiempo o para evaluar expresiones no-probabil√≠sticas con m√©todos probabil√≠sticos. Esta definici√≥n es mas restrictiva que la que dan otros autores para los cuales simulaci√≥n Montecarlo es cualquier simulaci√≥n que use n√∫meros aleatorios.

##**Aplicaci√≥n**

El m√©todo es aplicable a cualquier tipo de problema, ya sea estoc√°stico o determinista. A diferencia de los m√©todos num√©ricos que se basan en evaluaciones en N puntos en un espacio M-dimensional para producir una soluci√≥n aproximada, el m√©todo de Montecarlo tiene un error absoluto de la estimaci√≥n que decrece como $\frac {1}{\sqrt {N}}$ en virtud del teorema del l√≠mite central.

##**El teorema del l√≠mite central**

Tambi√©n conocido como teorema central del l√≠mite indica que, en condiciones muy generales, si $S_n$ es la suma de $n$ variables aleatorias independientes, con media conocida y varianza no nula pero finita, entonces la funci√≥n de distribuci√≥n de $S_n$ se aproxima a una distribuci√≥n normal. El teorema asegura que esto ocurre cuando la suma de estas variables aleatorias e independientes es lo suficientemente grande.

Si $X$ es una variable aleatoria tal que $ X\sim N(\mu ,\sigma ^{2})$ entonces su funci√≥n de densidad est√° dada por:

$f(x)={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}e^{\frac {-(x-\mu )^{2}}{2\sigma ^{2}}}$

Para $x\in\mathbb{R}$ donde $\mu$ denota la media y $\sigma ^{2}$ la varianza de la variable aleatoria $X$. En especial cuando $\mu = 0$ y $\sigma ^{2} = 1$ obtenemos:

$f(x)={\frac {1}{\sqrt {2\pi }}}e^{\frac {-x^{2}}{2}}$

es decir, la distribuci√≥n normal est√°ndar, denotada por $X\sim N(-1,1)$.

# **06. Caminata Aleatoria**
[[Contenidos]](#contents)

**Una caminata aleatoria es una ruta definida como una serie de pasos aleatorios.**

Las caminatas aleatorias pueden utilizarse para modelar fen√≥menos que ocurren en el mundo real, desde movimientos de mol√©culas en un gas hasta el comportamiento de un apostador que pasa un d√≠a en el casino. 
Una caminata aleatoria es una sucesi√≥n de variables aleatorias reales $(S)_{n=0}^{\infty}$ tal que $(S)_{n=0}$ y las variables aleatorias $(S_i - S_{i-1})_{i=1}^{\infty}$ son independientes e id√©nticamente distribuidas. Si $X_i = S_i-S_{i-1}$, entonces a la distribuci√≥n de $X_i$, que es independiente de $i$, y a la que daremos el nombre de distribuci√≥n del salto, gobierna a la caminata aleatoria y las caracter√≠sticas de esta √∫ltima se obtienen a trav√©s del estudio de dicha distribuci√≥n. 

Si $X_i$ tiene una distribuci√≥n Bernoulli sim√©trica de par√°metro $1/2$ se le conoce como caminata aleatoria simple y ciertas distribuciones relacionadas con esta sucesi√≥n se encuentran mediante m√©todos combinatorios; sin embargo, para el caso general, los m√©todos a utilizar generalmente son mas complicados y el objetivo de su estudio es expresar la soluci√≥n al problema planteado en t√©rminos de las caminatas aleatorias a trav√©s de la funci√≥n de distribuci√≥n. A veces tal soluci√≥n es muy complicada y no es de utilidad practica, por lo que adem√°s de la soluci√≥n exacta, generalmente buscamos expresiones aproximadas mas sencillas. Este enfoque se puede utilizar aun cuando la soluci√≥n exacta a nuestro problema nos sea desconocida.
